### Базовые операции над массивами

Важно, что в numpy операции над массивами происходят векторизованно для всех элементов массива сразу, причем вычисления столь же эффективны, как в языке C. 

```python
x = np.arange(10)
y = np.arange(15, 5, -1)
x, y
```

```python
# поэлементное сложение
x + y
```

```python
# поэлементное умножение
x * y
```

```python
# поэлементный sin
np.sin(x)
```

```python
# сумма элементов в массиве
np.sum(x)
```

```python
# произведение элементов в массиве
np.prod(x)
```

```python
# нахождение максимума
np.amax(x * y)
```

```python
# нахождение аргмаксимума
np.argmax(x * y)
```

```python
# поэлементное сравнение
x > 5
```

```python
# выполнение условия для всех элементов
np.all(x > 5)
```

```python
# выполнение условия хотя бы для одного элемента
np.any(x > 5)
```

```python
# индексы элементов, где выполняется условие
np.where(x > 5)
```

```python
# разность соседних элементов
np.diff(x)
```

### Изменение размера массивов

numpy-массивы всегда лежат в памяти последовательно, одним куском, поэтому у массива можно изменять размер практически бесплатно.

```python
x = np.array([[1, 2, 3], [4, 5, 6]])
x, x.shape
```

```python
# вытянуть в вектор
x = x.ravel()
x, x.shape
```

```python
# добавить фиктивную размерность (размера 1)
x = x[:, np.newaxis]
x, x.shape
```

```python
# другой способ добавить фиктивную размерность
x = x.ravel()
x = x.reshape(-1, 1)
x, x.shape
```

```python
# изменить размерность, идем по строкам
x = x.reshape(3, 2)
x, x.shape
```

### Операции над многомерными массивами

```python
x = np.array([[1, 2], [2, 1], [2, 3]])
```

```python
# транспонирование
x.T
```

```python
# матричное умножение
y = np.array([[0, 1, 0], [1, 0, 1]])
x.dot(y)
```

```python
# матричное умножение
res = np.dot(x, y)
res, res.shape
```

```python
# взятие диагонали 
np.diag(res)
```

```python
# заменить элементы над диагональю на 0
np.tril(res)
```

```python
# заменить элементы под диагональю на 0
np.triu(res)
```

* после применения функции получится массив, в котором будет отсутствовать размерность указанная в axis, но остальные размерности будут иметь такие же значения

```python
x, x.shape
```

```python
# максимум в каждой строке
np.amax(x, axis=1)
```

```python
# сумма по столбцам
np.sum(x, axis=0)
```

```python
# сумма по строкам
np.sum(x, axis=1)
```

### Broadcasting


```python
x = np.arange(10)
y = np.arange(5)
```

Можно складывать только массивы одного размера:

```python
x + y
```

Или почти одного...

```python
x[:, np.newaxis] + y[np.newaxis, :]
```

***Обязательно к прочтению!*** Подробное описание: http://docs.scipy.org/doc/numpy/user/basics.broadcasting.html.

Бродкастинг - автоматическое расширение размерности и размеров массивов, при совершении операций (сложение, умножение и подобные) над массивами с разными размерами или размерностями, при условии, что они совместимы с правилами бродкастинга.

Проверка на совместимость по размерам
1. Смотрим попарные размеры массивов, начиная с правого края. Если слева не хватает размерности, то она автоматически расширяется единицей. 
2. Размерности совместимы, если:
  1. у них одинаковые размерности
  2. одна размерность имеет размер 1 - тогда элементы соответсвующего массива "копируются" для реализации операции (физически копирования не происходит).

Пример:

Даны x - массив размера [8, 4, 6, 1], y - массив размера [4, 1, 5].

Сравнимаем размерности, начиная с правого края:
1. 1 (4 размерность x) и 5 (3 размерность y) - совместимы
2. 6 (3 размерность x) и 1 (2 размерность y) - совместимы 
3. 4 (2 размерность x) и 4 (1 размерность y) - совместимы 
4. 8 (1 размерность x) и 1 (добавленная размерность y) - совместимы 

```python
x = np.zeros((8, 4, 6, 1))
y = np.ones((4, 1, 5))

(x + y).shape
```

### View и копирование

```python
x = np.arange(10)
x
```

```python
y = x
y.shape = (2, 5)
x
```

View ссылается на те же данные, но позволяет задать другие размерности массива.

```python
x = np.arange(10)
v = x.view()
v.shape = (2, 5)
x
```

```python
v[0, 0] = 100
x
```

Если нужно получить копию массива, чтобы не портить переданные данные, пользуйтесь функцией copy()

```python
x = np.arange(10)
y = x.copy()
y[:] = 0
x
```
